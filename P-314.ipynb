```python
corr_table = df.corr(method="pearson") #get variables the correlation

###plotting the correlation table using seaborn
plt.figure(figsize=(16,4))
sns.heatmap(corr_table, annot=True, fmt=".1g")
plt.title("Correlation Heatmap between variables")
plt.show() #display the plot
```

```python
sns.regplot(data=df, y="loudness", x="energy", color="b")
plt.title("regression plot of the correlation of energy vs loudness")
plt.show()
```

```python
df.hist()
```

```python
plt.figure(figsize=(20,5))
sns.stripplot(x='track_genre', y='energy',hue='time_signature', data=df)
plt.ylabel('song')
plt.show()
```

# Top 10 Album

```python
trending_album=df[['album_name','popularity']].sort_values(by='popularity',ascending=False)[:10]
trending_album
```

### top 10 artist

```python
popular_artists = df.groupby("artists").count().sort_values(by='popularity', ascending=False)['popularity'][:10]
popular_artists
```

```python
ax = popular_artists.plot.barh()

for index, value in enumerate(popular_artists):
    ax.text(value, index, str(value))

plt.title("Top 10 artists")
plt.xlabel("Popularity Count")
plt.ylabel("Artists")
plt.show()
```

```python
longest_song=df[['track_name','duration_ms']].sort_values(by='duration_ms',ascending=False)[:10]
longest_song
```

```python
sns.barplot(x='duration_ms',y='track_name',data=longest_song)

plt.title("Longest Songs Duration")
plt.show()
```

## Top Five trending Genre

```python
trending_genre=df[['track_genre','popularity']].sort_values(by='popularity',ascending=False)[:5]
trending_genre
```

```python
sns.barplot(x='track_genre',y='popularity',data=trending_genre)
plt.title('Top 5 Genre')
plt.show()
```

## Top 5 Danceble songs

```python
danceble_songs=df[["track_name", "artists", "danceability"]].sort_values(by="danceability", ascending=False)[:5]
danceble_songs
```

```python
plt.pie(x="danceability", data=danceble_songs, autopct='%1.2f%%', labels=danceble_songs.track_name)
plt.title("Top 5 Most Danceable Songs")
plt.show()
```
